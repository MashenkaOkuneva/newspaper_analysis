In this notebook, we visualize the relationship between the number of topics in an LDA model and the corresponding perplexity on test data. We plot perplexity values for models with 10, 50, 100, 150, 200, and 250 topics, with a vertical line highlighting the selection of 200 topics.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from CSV files
perplexity_10_100 = pd.read_csv('perplexity_k_10_100.csv', header=None)
perplexity_150_250 = pd.read_csv('perplexity_k_150_250.csv', header=None)

# Combine the perplexity values and corresponding topic numbers
perplexity_values = perplexity_10_100[0].tolist() + perplexity_150_250[0].tolist()
num_topics = [10, 50, 100, 150, 200, 250]

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(num_topics, perplexity_values, marker='o', label='Perplexity')

# Label the axes
plt.xlabel('Number of Topics')
plt.ylabel('Perplexity')

# Add title
#plt.title('Perplexity vs Number of Topics')

# Add vertical line at 200
plt.axvline(x=200, color='red', linestyle='--', label='Selected: 200 Topics')

# Add grid and legend
plt.grid(True)
plt.legend()

# Save the plot to PNG and EPS files
plt.savefig('perplexity_plot.png', format='png')
plt.savefig('perplexity_plot.eps', format='eps')

# Show the plot
plt.show()
```
